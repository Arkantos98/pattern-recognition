```python
from tabulate import tabulate
from beautifultable import BeautifulTable
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
#'Hobby'

hobby_list = df['Hobby'].value_counts()
hobby_list_idxs = hobby_list.index.tolist()
hl = df['Hobby'].count()

hobby_table = BeautifulTable()
for h,v in zip(hobby_list_idxs, hobby_list):
    hobby_table.append_row([h, v, str(100*v/hl) + "%"])

print(tabulate(hobby_table, headers=["Por hobby", "Participantes", "Porcentaje"]))  
```

```python
#'OpenSource'

osource_list = df['OpenSource'].value_counts()
osource_list_idxs = osource_list.index.tolist()
ol = df['OpenSource'].count()

osource_table = BeautifulTable()
for h,v in zip(osource_list_idxs, osource_list):
    osource_table.append_row([h, v, str(100*v/ol) + "%"])

print(tabulate(osource_table, 
               headers=["Proyectos Open Source", "Participantes", "Porcentaje"]))  
```

### <span style="color:#313f9e; font-weight: bold;">Verificación de la calidad de los datos:</span>
<div style="margin-top: 6px; border: 1px solid #cfcfcf; padding: 8px 12px; border-radius:2px; background-color:#f7f7f7;">

El problema evidente, en este conjunto de datos, es el de valores faltantes. Para determinar la magnitud del problema, calculamos el porcentaje de valores faltantes en cada columna: Un porcentaje superior al 5% será un problema serio.
</div>

```python
table = BeautifulTable()
for col in df.describe(include='all'):
    table.append_row([col, 100*sum(pd.isnull(df[col]))/df.shape[0]])

    
    
print(tabulate(table, headers=["Columna", "Valores Faltantes (%)"]))    
```

<div style="margin-top: 6px; border: 1px solid #cfcfcf; padding: 8px 12px; border-radius:2px; background-color:#f7f7f7;">
Dado que los datos son generados de una encuesta con reactivos de opción múltiple, no se espera que haya valores atípicos o fuera de rango.  
<br>
Para detectar otros tipos de patrones, como variables altamente correlacionadas, requerimos primero hacer la recodificación de variables, por lo que dejamos ese análisis para la fase de preparación de los datos.
