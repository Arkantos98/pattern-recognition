```python
import pydotplus # brew install graphviz, pip install pydotplus
from io import StringIO
from IPython.display import Image, display  

heroesData = pd.read_csv("Data sets/superHeoes.csv")
display(heroesData)

heroesData = pd.read_csv("Data sets/superHeoes2.csv")
featuresHDf = heroesData[list(['Power origin', 'Alter ego', 'Team player', 'Human'])]
classesHDf = heroesData[list(['Class'])]
clfH = tree.DecisionTreeClassifier()
clfH = clfH.fit(featuresHDf, classesHDf)

dotfileH = StringIO()
tree.export_graphviz(clfH, out_file=dotfileH, filled=False, class_names=["A", "B"],
                     feature_names=['Power origin', 'Alter ego', 'Team player', 'Human'], 
                     rounded=True)
graph = pydotplus.graph_from_dot_data(dotfileH.getvalue())
display(Image(graph.create_png()))
```
