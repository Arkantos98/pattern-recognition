## Técnicas de clustering: Comparación de técnicas de clustering

Este resumen de técnicas de *clustering* está basado y utiliza gran parte del código de [Comparing different clustering algorithms on toy datasets](http://scikit-learn.org/stable/auto_examples/cluster/plot_cluster_comparison.html). El objetivo es mostar el comportamiento de algunas de las técnicas de clustering sobre diversos conjuntos de datos sintéticos. 

<font color="red" style="font-size:20px">NOTA:</font> Puesto que una parte importante del uso de las técnicas de Reconocimiento de patrones es la elección y ajuste de difefrentes parámetros, los resultados no son de ninguna manera concluyentes sobre la calidad de los diferentes métodos.

```python
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import cluster, datasets
from sklearn.neighbors import kneighbors_graph
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
```

### Casos de estudio:

```python
np.random.seed(0)

# Generate datasets. We choose the size big enough to see the scalability
# of the algorithms, but not too big to avoid too long running times
n_samples = 1500

databases_names = [
    'Círculos con ruido', 'Lunas con ruido', 'Blobs', 'Curva S con ruido',
    'Rollo suizo y blobs', 'Factoextra']

plt.figure(figsize=(12, 8))
plt.subplots_adjust(wspace=.25, hspace=.25)

noisy_circles = datasets.make_circles(n_samples=n_samples, factor=.5, noise=.05)
plt.subplot(2, 3, 1).set_title(databases_names[0])
plt.plot(noisy_circles[0][:, 0], noisy_circles[0][:, 1], '.', 
         markeredgecolor='0', markerfacecolor='0')

noisy_moons = datasets.make_moons(n_samples=n_samples, noise=.05)
plt.subplot(2, 3, 2).set_title(databases_names[1])
plt.plot(noisy_moons[0][:, 0], noisy_moons[0][:, 1], '.', 
         markeredgecolor='0', markerfacecolor='0')

blobs = datasets.make_blobs(n_samples=n_samples, random_state=8)
plt.subplot(2, 3, 3).set_title(databases_names[2])
plt.plot(blobs[0][:, 0], blobs[0][:, 1], '.', markeredgecolor='0', markerfacecolor='0')

baux = datasets.make_blobs(n_samples=500, random_state=5)
noisy_scurve = datasets.make_s_curve(n_samples=1000, noise=.05)
#remove z component (x,z,y)
noisy_scurve = (np.delete(noisy_scurve[0], 1, 1) + 5, noisy_scurve[1])
noisy_scurve = (np.append(noisy_scurve[0], baux[0] / 3 + 5, axis = 0), 
                   np.append(noisy_scurve[1], baux[1]))
plt.subplot(2, 3, 4).set_title(databases_names[3])
plt.plot(noisy_scurve[0][:, 0], noisy_scurve[0][:, 1], '.', 
         markeredgecolor='0', markerfacecolor='0')

noisy_swissroll = datasets.make_swiss_roll(n_samples=1000, noise=0.5)
#remove z component (x,z,y)
noisy_swissroll = (np.delete(noisy_swissroll[0] / 2 + 10, 1, 1), noisy_swissroll[1])
noisy_swissroll = (np.append(noisy_swissroll[0], baux[0], axis = 0), 
                   np.append(noisy_swissroll[1], baux[1]))
plt.subplot(2, 3, 5).set_title(databases_names[4])
plt.plot(noisy_swissroll[0][:, 0], noisy_swissroll[0][:, 1], '.', 
         markeredgecolor='0', markerfacecolor='0')

multishapes = (pd.read_csv("Data sets/Factoextra/multishapes_raw.csv").values, None)
plt.subplot(2, 3, 6).set_title(databases_names[5])
plt.plot(multishapes[0][:, 0], multishapes[0][:, 1], 'o', 
         markeredgecolor='0', markerfacecolor='0', markersize=3)

plt.show()

datasets = [noisy_circles, noisy_moons, blobs, noisy_scurve, noisy_swissroll, multishapes]
```

### Dendrogramas:

```python
plt.figure(figsize=(12, 5))
plt.subplots_adjust(hspace=1.0)

for i_dataset, dataset in enumerate(datasets):
    X, y = dataset
    
    # normalize dataset for easier parameter selection
    X = StandardScaler().fit_transform(X)
    
    plt.subplot(2, 3, i_dataset + 1)
    plt.subplot(2, 3, i_dataset + 1).set_title(databases_names[i_dataset])
    Z = linkage(X, 'complete')
    dendrogram(Z, truncate_mode='lastp', p=6, show_leaf_counts=True, 
               leaf_rotation=75, leaf_font_size=14.)
```

### Aplicación de otros métodos de *clustering*

```python
colors = np.array([x for x in 'bgrcmykbgrcmykbgrcmykbgrcmyk'])
colors = np.hstack([colors] * 20)

clustering_names = [
    'KMeans', 'DBSCAN', 'AffinityPropagation', 'MeanShift',
    'SpectralClustering', 'Ward', 'AgglomerativeClustering',
    'Birch']

plt.figure(figsize=(len(clustering_names) * 2 + 3, 27))
plt.subplots_adjust(hspace=0.35)

plot_num = 1

num_clusters = [2, 2, 3, 3, 3, 5]
for i_dataset, dataset in enumerate(datasets):
    X, y = dataset
    
    # normalize dataset for easier parameter selection
    X = StandardScaler().fit_transform(X)

    # create k-means estimator
    k_means = cluster.KMeans(n_clusters=num_clusters[i_dataset], init='random')

    # create DBSCAN estimator
    dbscan = cluster.DBSCAN(eps=.2, min_samples=6)

    # create AffinityPropagation estimator
    affinity_propagation = cluster.AffinityPropagation(damping=.8, preference=-50)

    # create MeanShift estimator
    bandwidth = cluster.estimate_bandwidth(X, quantile=0.3)
    ms = cluster.MeanShift(bandwidth=bandwidth, bin_seeding=True)
    
    # create SpectralClustering estimator
    spectral = cluster.SpectralClustering(n_clusters = num_clusters[i_dataset],
                                          eigen_solver='arpack',
                                          affinity="rbf")

    # create Ward estimator
    ward = cluster.AgglomerativeClustering(n_clusters = num_clusters[i_dataset], linkage='ward')
    # create Ward estimator
    average_linkage = cluster.AgglomerativeClustering(
        linkage="average", affinity="cityblock", n_clusters=num_clusters[i_dataset])

    # create Birch estimator
    birch = cluster.Birch(n_clusters = num_clusters[i_dataset])
    
    clustering_algorithms = [
        k_means, dbscan, affinity_propagation, ms, spectral, ward, average_linkage, birch
    ]

    alg = 0
    for name, algorithm in zip(clustering_names, clustering_algorithms):
        # predict cluster memberships
        
        t0 = time.time()
        algorithm.fit(X)
        t1 = time.time()
        if hasattr(algorithm, 'labels_'):
            y_pred = algorithm.labels_.astype(np.int)
        else:
            y_pred = algorithm.predict(X)

        # plot
        plot_num = i_dataset + alg * 6 + 1
        plt.subplot(len(clustering_algorithms), 6, plot_num)
        if i_dataset == 0:
            plt.title(name, size=18)
        plt.scatter(X[:, 0], X[:, 1], color=colors[y_pred].tolist(), s=10)
        
        if hasattr(algorithm, 'cluster_centers_'):
            centers = algorithm.cluster_centers_
            center_colors = colors[:len(centers)]
            plt.scatter(centers[:, 0], centers[:, 1], s=100, c=center_colors)
        plt.xlim(-2, 2)
        plt.ylim(-2, 2)
        plt.xticks(())
        plt.yticks(())
        plt.text(.99, -.12, ('%.2fs' % (t1 - t0)).lstrip('0'),
                 transform=plt.gca().transAxes, size=15,
                 horizontalalignment='right')
        alg += 1
```
